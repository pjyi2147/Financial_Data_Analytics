```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
# load q3 data from DA1.xlsx
merge_data = pd.read_excel('../DA2_data.xlsx', sheet_name='q3')
```

```python
merge_data['lnSize'] = np.log(merge_data['PRC'] * merge_data['SHROUT'] * 1000)
merge_data['bk2mkt'] = merge_data['CEQQ'] * 1000000 / (merge_data['PRC'] * merge_data['SHROUT'] * 1000)
merge_data['eP'] = merge_data['IBQ'] / (merge_data['PRC'] * merge_data['SHROUT']) * 1000
merge_data['eP_2'] = merge_data['EPSPXQ'] / merge_data['PRC']

# test = merge_data[merge_data['PRC'].notnull()]
# test = test[test['EPSPXQ'].notnull()]
# print(test['PERMNO'].count())

# test2 = merge_data[merge_data['IBQ'].notnull()]
# test2 = test2[test2['PRC'].notnull()]
# test2 = test2[test2['SHROUT'].notnull()]
# print(test2['PERMNO'].count())

# print(merge_data[['MONTHID', 'PERMNO', 'eP']].head())

# print(merge_data[['lnSize']].describe())
# print(merge_data[['bk2mkt']].describe())
# print(merge_data[['eP']].describe())
# print(merge_data[['eP_2']].describe())
```

```python
# plot the histogram of the lnSize
plt.hist(merge_data['lnSize'], bins=100)
plt.title('Histogram of lnSize')
plt.xlabel('lnSize')
plt.ylabel('Frequency')
plt.show()
```

```python
# plot the histogram of bk2mkt
plt.hist(merge_data['bk2mkt'], bins=100)
plt.title('Histogram of bk2mkt')
plt.xlabel('bk2mkt')
plt.ylabel('Frequency')
plt.show()
```
