```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data from output excel file
q4_data = pd.read_excel('../DA2_data.xlsx', sheet_name='q4')
```

```python
q6_cols = ['lnSize', 'bk2mkt', 'eP']

# for each MONTHID, winsorize lnSize, bk2mkt, and eP based on 3 standard deviations
# and indicate the winsorized data
for col in q6_cols:
    q4_data[col+'_W'] = q4_data.groupby('MONTHID')[col].transform(lambda x: x.clip(lower=x.mean() - 3*x.std(), upper=x.mean() + 3*x.std()))
    q4_data[col+'_W_flag'] = np.where(q4_data[col+'_W'] == q4_data[col], False, True)
    q4_data[col+'_T'] = q4_data.groupby('MONTHID')[col].transform(lambda x: x.where(x.between(x.mean()-3*x.std(), x.mean()+3*x.std()), np.nan))
    q4_data[col+'_T_flag'] = np.where(q4_data[col+'_T'] == q4_data[col], False, True)
```

```python
# count the number of winsorized and truncated data
print('Number of total outliers of lnSize:', q4_data['lnSize_W_flag'].sum())
print('Number of total outliers of bk2mkt:', q4_data['bk2mkt_W_flag'].sum())
print('Number of total outliers of eP:', q4_data['eP_W_flag'].sum())
print()

# report statistics for winsorized data
print('Statistics for winsorized data for lnSize:')
print('Mean: ' + str(q4_data['lnSize_W'].mean().round(5)))
print('Median: ' + str(q4_data['lnSize_W'].median().round(5)))
print('Standard deviation: ' + str(q4_data['lnSize_W'].std().round(5)))
print('Min: ' + str(q4_data['lnSize_W'].min().round(5)))
print('Max: ' + str(q4_data['lnSize_W'].max().round(5)))
print('1% percentile: ' + str(q4_data['lnSize_W'].quantile(0.01).round(5)))
print('99% percentile: ' + str(q4_data['lnSize_W'].quantile(0.99).round(5)))
print()

# report statistics for truncated data
print('Statistics for truncated data for lnSize:')
print('Mean: ' + str(q4_data['lnSize_T'].mean().round(5)))
print('Median: ' + str(q4_data['lnSize_T'].median().round(5)))
print('Standard deviation: ' + str(q4_data['lnSize_T'].std().round(5)))
print('Min: ' + str(q4_data['lnSize_T'].min().round(5)))
print('Max: ' + str(q4_data['lnSize_T'].max().round(5)))
print('1% percentile: ' + str(q4_data['lnSize_T'].quantile(0.01).round(5)))
print('99% percentile: ' + str(q4_data['lnSize_T'].quantile(0.99).round(5)))
print()

# report statistics for winsorized data
print('Statistics for winsorized data for bk2mkt:')
print('Mean: ' + str(q4_data['bk2mkt_W'].mean().round(5)))
print('Median: ' + str(q4_data['bk2mkt_W'].median().round(5)))
print('Standard deviation: ' + str(q4_data['bk2mkt_W'].std().round(5)))
print('Min: ' + str(q4_data['bk2mkt_W'].min().round(5)))
print('Max: ' + str(q4_data['bk2mkt_W'].max().round(5)))
print('1% percentile: ' + str(q4_data['bk2mkt_W'].quantile(0.01).round(5)))
print('99% percentile: ' + str(q4_data['bk2mkt_W'].quantile(0.99).round(5)))
print()

# report statistics for truncated data
print('Statistics for truncated data for bk2mkt:')
print('Mean: ' + str(q4_data['bk2mkt_T'].mean().round(5)))
print('Median: ' + str(q4_data['bk2mkt_T'].median().round(5)))
print('Standard deviation: ' + str(q4_data['bk2mkt_T'].std().round(5)))
print('Min: ' + str(q4_data['bk2mkt_T'].min().round(5)))
print('Max: ' + str(q4_data['bk2mkt_T'].max().round(5)))
print('1% percentile: ' + str(q4_data['bk2mkt_T'].quantile(0.01).round(5)))
print('99% percentile: ' + str(q4_data['bk2mkt_T'].quantile(0.99).round(5)))
print()

# report statistics for winsorized data
print('Statistics for winsorized data for eP:')
print('Mean: ' + str(q4_data['eP_W'].mean().round(5)))
print('Median: ' + str(q4_data['eP_W'].median().round(5)))
print('Standard deviation: ' + str(q4_data['eP_W'].std().round(5)))
print('Min: ' + str(q4_data['eP_W'].min().round(5)))
print('Max: ' + str(q4_data['eP_W'].max().round(5)))
print('1% percentile: ' + str(q4_data['eP_W'].quantile(0.01).round(5)))
print('99% percentile: ' + str(q4_data['eP_W'].quantile(0.99).round(5)))
print()

# report statistics for truncated data
print('Statistics for truncated data for eP:')
print('Mean: ' + str(q4_data['eP_T'].mean().round(5)))
print('Median: ' + str(q4_data['eP_T'].median().round(5)))
print('Standard deviation: ' + str(q4_data['eP_T'].std().round(5)))
print('Min: ' + str(q4_data['eP_T'].min().round(5)))
print('Max: ' + str(q4_data['eP_T'].max().round(5)))
print('1% percentile: ' + str(q4_data['eP_T'].quantile(0.01).round(5)))
print('99% percentile: ' + str(q4_data['eP_T'].quantile(0.99).round(5)))
```
