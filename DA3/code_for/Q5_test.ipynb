```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import pearsonr, ttest_ind
```

```python
q5_data = pd.read_excel('../DA3_data.xlsx', sheet_name="Q5")
```

```python
print(q5_data[['BETA', 'BETA_W', 'IVOL', 'IVOL_W', 'MOM', 'MOM_W', 'RET_L']].sum())
```

```python
cols = pd.DataFrame(columns=['BETA_W', 'IVOL_W', 'MOM_W', 'lnSize_W', 'bk2mkt_W', 'eP_W', 'RET_L'])
```

```python
display(round(q5_data[cols.columns].corr(), 6))
pvalues = cols.transpose().join(cols, how='outer')
pvalues2 = cols.transpose().join(cols, how='outer')
for r in cols:
    for c in cols:
        tmp = q5_data[q5_data[r].notnull() & q5_data[c].notnull()]
        _, p = pearsonr(tmp[r], tmp[c])
        pvalues[r][c] = round(p, 6)
        model = sm.OLS(q5_data[r], sm.add_constant(q5_data[c]), missing='drop').fit()
        pvalues2[r][c] = round(model.pvalues[c], 6)
display(pvalues)
display(pvalues2)
```
